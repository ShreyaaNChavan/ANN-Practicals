- Generate ANDNOT function using McCulloch-Pitts neural net.

```python
# AND-NOT using Perceptron
def perceptron(A, B):
    w1 = 1
    w2 = -1
    bias = -0.5

    # Step 1: Calculate weighted sum
    z = (w1 * A) + (w2 * B) + bias
    print("Weighted Sum (z):", z)

    # Step 2: Activation
    if z >= 0:
        return 1
    else:
        return 0
```

```python
# User Input
n = int(input("Enter number of test cases: "))

for i in range(n):
    print(f"\nTest Case {i+1}")
    A, B = map(int, input("Enter A and B (0 or 1): ").split())

    output = perceptron(A, B)

    print("Final Output:", output)
```
