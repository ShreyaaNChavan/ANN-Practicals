```python
import numpy as np

# Activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def derivative(x):
    return x * (1 - x)

# XOR input and output
X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])

Y = np.array([[0],
              [1],
              [1],
              [0]])

# Initialize weights
np.random.seed(0)
W1 = np.random.rand(2, 2)
W2 = np.random.rand(2, 1)

# Training
for i in range(5000):
    # Forward pass
    H = sigmoid(np.dot(X, W1))
    O = sigmoid(np.dot(H, W2))

    # Error
    error = Y - O

    # Print step-by-step for first few iterations
    if i < 5:
        print(f"\nEpoch {i+1}")
        print("Predicted Output:\n", O)
        print("Actual Output:\n", Y)
        print("Error (Y - O):\n", error)
        print("Mean Error:", np.mean(np.abs(error)))

    # Backpropagation
    dO = error * derivative(O)
    dH = dO.dot(W2.T) * derivative(H)

    # Update weights
    W2 += H.T.dot(dO)
    W1 += X.T.dot(dH)

# Final outputs
print("\n--- After Training ---")
print("Final Predicted Output:\n", O)

# User input
a = int(input("\nEnter first input (0 or 1): "))
b = int(input("Enter second input (0 or 1): "))

# Prediction
inp = np.array([[a, b]])
h = sigmoid(np.dot(inp, W1))
o = sigmoid(np.dot(h, W2))

print("Output:", round(o[0][0]))
```
