- Perceptron to classify ASCII digits (0-9) as Even (1) or Odd (0)


```python
import random
```

```python
# Step 1: Prepare Training Data
training_data = []

for digit in range(10):
    ascii_value = ord(str(digit))
    binary = format(ascii_value, '08b')
    inputs = [int(bit) for bit in binary]

    target = 1 if digit % 2 == 0 else 0
    training_data.append((inputs, target))
```

```python
# Step 2: Initialize weights and bias
weights = [random.uniform(-1, 1) for _ in range(8)]
bias = random.uniform(-1, 1)

learning_rate = 0.1
epochs = 20
```

```python
# Step 3: Training
for epoch in range(epochs):
    total_error = 0

    for inputs, target in training_data:

        z = sum(weights[i] * inputs[i] for i in range(8)) + bias
        output = 1 if z >= 0 else 0

        error = target - output
        total_error += abs(error)

        for i in range(8):
            weights[i] += learning_rate * error * inputs[i]
        bias += learning_rate * error

    # Print error and parameters after each epoch
    print(f"Epoch {epoch + 1}")
    print("Error:", total_error)
    print("Weights:", weights)
    print("Bias:", bias)
    print("-" * 40)

    if total_error == 0:
        break

print("\nTraining Completed")
print("Final Weights:", weights)
print("Final Bias:", bias)
```

```python
# Step 4: User Testing Section
while True:
    print("\nChoose an option:")
    print("1. Enter your own digit")
    print("2. Test random digit")
    print("3. Exit")

    choice = input("Enter choice (1/2/3): ")

    if choice == "1":
        user_input = input("Enter a digit (0-9): ")

        if not user_input.isdigit() or not (0 <= int(user_input) <= 9):
            print("Invalid input! Please enter a single digit (0-9).")
            continue

        digit = int(user_input)

    elif choice == "2":
        digit = random.randint(0, 9)
        print("Random Digit Generated:", digit)

    elif choice == "3":
        print("Exiting Program...")
        break

    else:
        print("Invalid choice!")
        continue

    # Convert digit to binary ASCII
    ascii_value = ord(str(digit))
    binary = format(ascii_value, '08b')
    inputs = [int(bit) for bit in binary]

    # Prediction
    z = sum(weights[i] * inputs[i] for i in range(8)) + bias
    prediction = 1 if z >= 0 else 0

    result = "Even" if prediction == 1 else "Odd"
    print("Prediction: Digit", digit, "is", result)
```
